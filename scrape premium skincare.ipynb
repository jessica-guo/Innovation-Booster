```python
def get_features(product_url,driver):
    dict_ = dict()
    dict_['Source'] = 'Dermstore'
    dict_['Link'] = product_url
    try:
        Name_list = driver.find_elements_by_xpath("//h1[@class='prod-name visible-xs']")
        Name = Name_list[0].text
        dict_['Name'] = Name
    except:
        pass
    try:
        Price_list = driver.find_elements_by_xpath("//span[@class='formGroup=DS--label']/strong")
        Price = Price_list[0].text 
        dict_['Price'] = Price
    except:
        pass
    try:
        Intro_list = driver.find_elements_by_xpath("//div[@class='panel-body']/p")
        Intro = Intro_list[1].text
        dict_['Intro'] = Intro
    except:
        pass
    try:
        Key_list = driver.find_elements_by_xpath("//div[@class='panel-body']/ul")
        Key_Ingredients = Key_list[0].text
        dict_['Key_Ingredients'] =  Key_Ingredients
    except:
        pass
    try:
        Key_Benefits = Key_list[1].text
        dict_['Key_Benefits'] =  Key_Benefits
    except:
        pass
    try:
        detail_list = driver.find_elements_by_xpath("//div[@class='prd']")
        Ideal_for_concern = detail_list[1].text.split(': ')[1]
        dict_['Ideal_for_concern'] =  Ideal_for_concern
    except:
        pass
    try:
        Ideal_for_skintype = detail_list[2].text.split(': ')[1]
        dict_['Ideal_for_skintype'] =  Ideal_for_skintype
    except:
        pass
    try:
        preference = detail_list[3].text.split(': ')[1]
        dict_['preference'] =  preference
    except:
        pass
    try:
        Rating_list = driver.find_elements_by_xpath("//div[@class='statbar--overview--rating']/sup[@id='TOTAL']")
        Rating = Rating_list[0].text
        dict_['Rating'] =  Rating
    except:
        pass
    return dict_
```
